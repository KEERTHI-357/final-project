```python
# ===============================
# Warehouse Space Optimization
# ===============================

# ✅ STEP 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# ✅ STEP 2: Load Dataset
df = pd.read_csv('/content/drive/MyDrive/warehouse_products.csv')
print("Initial Data Shape:", df.shape)
df.head()

# ✅ STEP 3: Data Cleaning
# Identify numeric & categorical columns
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns

# Handle missing numeric values
num_imputer = SimpleImputer(strategy='mean')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Handle missing categorical values
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# ✅ Remove outliers using Z-Score
from scipy.stats import zscore
z_scores = np.abs(zscore(df[num_cols]))
df = df[(z_scores < 3).all(axis=1)]
print("After Outlier Removal:", df.shape)

# ✅ STEP 4: Feature Scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[num_cols])

# ✅ STEP 5: Optimal Clusters (Elbow )
inertia = []
sil_scores = []
K = range(2, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(scaled_features, kmeans.labels_))

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(K, inertia, 'bo-')
plt.title("Elbow Method (Inertia)")
plt.xlabel("K")
plt.ylabel("Inertia")

plt.subplot(1,2,2)
plt.plot(K, sil_scores, 'ro-')
plt.title("Silhouette Score")
plt.xlabel("K")
plt.ylabel("Score")
plt.show()

# ✅ STEP 6: K-Means Clustering
best_k = 4  # choose based on plots
kmeans = KMeans(n_clusters=best_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# ✅ STEP 7: Dimensionality Reduction (PCA)
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)
df['PCA1'], df['PCA2'] = pca_features[:,0], pca_features[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='tab10', data=df)
plt.title("Clusters Visualized (PCA)")
plt.show()

# ✅ STEP 8: Cluster Profiling
cluster_profile = df.groupby('Cluster')[num_cols].mean()
print("\nCluster Profiles:")
print(cluster_profile)

# ✅ Save Cleaned Data for Streamlit
df.to_csv('/content/warehouse_cleaned.csv', index=False)
print("\nCleaned dataset saved as 'warehouse_cleaned.csv'")
```
